```r
library(ggplot2)
```

## Comparison between OPUS Impact rule and Apriori

```r
df<- read.csv('aggregate summary 1.3.csv')
```

### Rule time comparison

#### Run time with rank by number of transactions in dataset 

```r
# df2$dose <- as.numeric(as.vector(df2$dose))
ggplot(data=df, aes(x=rank.by.number.for.transaction, y=run.time, group=algorithm, color=algorithm)) +
  geom_line() + geom_point() + 
  theme_minimal()
```

### Number of rules comparison

```r
ggplot(data=df, aes(x=rank.by.number.for.transaction, y=number.of.rule, group=algorithm, color=algorithm)) +
  geom_line() + geom_point() +
  theme_minimal()
```

## Comparison between OPUS Impact rule and productive impact rule

```r
df<- read.csv('summary 1.7.csv')
df$Non_productivity_rate = (df$IR.MU - df$IR.Mu.T.Pro)/df$IR.I
order.Non_productivity_rate = order(df$Non_productivity_rate)
df$rank <- NA
df$rank[order.Non_productivity_rate] <- 1:nrow(df)
ggplot(data=df, aes(x=rank, y=Non_productivity_rate)) +
  geom_line() + geom_point() + geom_text(aes(label=dataset),hjust=0, vjust=1,size=3) +
  theme_minimal()
```

```r
df<- read.csv('summary 1.7.csv')
df$Non_productivity_rate = (df$IR.I - df$IR.I.T.Pro)/df$IR.I
order.Non_productivity_rate  = order(df$Non_productivity_rate)
df$rank <- NA
df$rank[order.Non_productivity_rate] <- 1:nrow(df)
ggplot(data=df, aes(x=rank, y=Non_productivity_rate)) +
  geom_line() + geom_point() + geom_text(aes(label=dataset),hjust=0, vjust=1,size=3) +
  theme_minimal()
```
